```python
# this imports the pandas library
import pandas as pd
import matplotlib.pyplot as plt
# gets iris dataset from the file.
irisData=pd.read_csv('Downloads/iris.csv')
```

```python
# a) How many tuples does the Iris data set have?
num_tuples = len(irisData)
print("Number of tuples: ", num_tuples)
```

```python
# b) How many features does the Iris data set have? Print a name list

# Subtracts 1 for the target column
num_features = len(irisData.columns) - 1
# excludes the target column
feature_names= irisData.columns[:-1]
print("Number of features: ", num_features)
print("Feature names: ", list(feature_names))
```

```python
# c) What's the data type for each feature?
#prints the data types
print("Data types:")
print(irisData.dtypes)
```

```python
# d) How many of each species do you have?
# prnts out the species number
print("Number of each species:")
print(irisData["class"].value_counts())
```

```python
# e) What's the mean, median, and quantiles and ranges (max-min) for each petal and sepal measurement?
measurements = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
print("Measurements for each petal:")
print(irisData[measurements].describe())
```

```python
# f) make a scatter plot of sepal_length vs sepal_width
irisData.plot(kind="scatter", x="sepal_length", y="sepal_width")
plt.title("Spela Length vs Sepal Width")

#prints the scatter plot graph
plt.show()
```

```python
# g) Create four individual histogram plots in a single figure, 
# where each plot contains one feature 
# (petal_width, petal_length, sepal_width, sepal_length).

irisData.hist(column=measurements, figsize=(10, 8))
plt.suptitle("Histogram of Iris measurements")
plt.show()
```

```python
# h) Make a boxplot of each petal and sepal measurement.
irisData.boxplot(column=measurements, by="class", figsize=(10,8))
plt.suptitle("Boxplots of Iris measurements ")
```

```python
# i) Make a scatter plot matrix to examine the correlation between each of the measurements
pd.plotting.scatter_matrix(irisData[measurements], figsize=(10,10))
plt.suptitle("Scatter Plot Matrix")
plt.show()
```
